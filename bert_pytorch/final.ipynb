```python
import re
filters=set('! \" # $ % & ( ) * + , - . / : € ; < = > ? @ [ \\  ] ^ _ ` { | } ~ « σ ।\\ ร'.split()+['\\x02'])
telugu_chars=['ం', 'అ','ఆ','ఇ','ఈ', 'ఉ','ఊ','ఋ','ౠ','ఎ','ఏ','ఐ','ఒ','ఓ','ఔ','అం','అః','క','ఖ','గ','ఘ',
'ఙ','చ','ఛ','జ','ఝ','ఞ','ట','ఠ','డ','ఢ','ణ','త','థ','ద','ధ','న','ప','ఫ','బ','భ','మ','య',
'ర','ల','వ','శ','ష','స','హ','ళ','క్ష','ఱ','ా', 'ి','ీ','ు','ూ', 'ృ','ె','ే','ై','ొ','ో','ౌ','్' ]
english_chars='a b c d e f g h i j k l m n o p q r s t u v w x y z'.split()
with open('/home/sriharshkamma/final/telugu_dump_ourdata.txt','r') as f:
    data=f.read()
data=data.split('\n')[:-1]
with open('/home/sriharshkamma/bert/multi_cased_L-12_H-768_A-12/vocab.txt','r') as f:
    bert_vocab=f.read()
bert_vocab=bert_vocab.split('\n')[:-1]
```

```python
counter={}
for i in data:
    for j in i.split():
        if j in counter:
            counter[j]+=1
        else:
            counter[j]=1
sorted_words_dc = sorted(counter, key=counter.get, reverse=True)
sorted_words_ic = sorted(counter, key=counter.get, reverse=False)
y_dc=[counter[i] for i in sorted_words_dc]
y_ic=[counter[i] for i in sorted_words_ic]
len(counter)
```

```python
req_pc_coverage=0.75
z=[]
temp=0
sum_all=sum(y_ic)
for i in range(len(y_ic)):
    temp+=y_ic[i]
    z.append(temp/(sum_all))
first_time=1
tel_wc=0
eng_wc=0
useful_vocab_en=[]
useful_vocab_te=[]
for i in range(len(z)):
    if z[i]>=1-req_pc_coverage:
        if first_time:
            print('total = ',i,'\nat % =',(100*i)/len(z),'\nnum_of words=',len(z)-i)
            first_time=0
        else:
            if sorted_words_ic[i][0] in telugu_chars:
                tel_wc+=1
                useful_vocab_te.append(sorted_words_ic[i])
            else:
                eng_wc+=1
                useful_vocab_en.append(sorted_words_ic[i])
c=0
for i in bert_vocab:
    if len(i)>0 and i[0] in telugu_chars:
        useful_vocab_te.append(i)
        c+=1
useful_vocab_en+=english_chars
print('english vocab = ',len(useful_vocab_en),'telugu vocab = ',len(useful_vocab_te))
```

```python
with open('/home/sriharshkamma/final/BERT-pytorch/data/text_to_build_corpus.small','w') as f:
    for i in set(useful_vocab_en+useful_vocab_te):
        f.write(i+'\n')
```
